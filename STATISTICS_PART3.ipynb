##THEORY QUESTIONS



1. What is hypothesis testing in statistics?  
- Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves testing an assumption (called a hypothesis) using sample statistics.  
Example:  
Suppose you want to test if a new drug is more effective than the old one. You collect sample data and perform a hypothesis test to validate the claim.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?  
- Null Hypothesis (H₀): A default statement that there is no effect or no difference.  
- Alternative Hypothesis (H₁ or Ha): A statement that contradicts the null and suggests an effect or difference.  
Example:  
H₀: The new teaching method has no effect.  
H₁: The new teaching method improves scores.

3. What is the significance level in hypothesis testing, and why is it important?  
- The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. Common values: 0.05, 0.01.  
- It determines the threshold for decision-making.  
Example:  
If α = 0.05, you accept a 5% chance of making a Type I error.

4. What does a P-value represent in hypothesis testing?  
- The P-value is the probability of obtaining test results at least as extreme as the observed results, assuming the null hypothesis is true.  
Example:  
If P-value = 0.03, it means there's a 3% chance the results happened under H₀.

5. How do you interpret the P-value in hypothesis testing?  
- If P-value ≤ α, reject the null hypothesis (evidence against H₀).  
- If P-value > α, fail to reject the null hypothesis (not enough evidence).  
Example:  
P = 0.02, α = 0.05 → Reject H₀  
P = 0.07, α = 0.05 → Fail to reject H₀

6. What are Type I and Type II errors in hypothesis testing?  
- Type I Error: Rejecting H₀ when it’s true (false positive).  
- Type II Error: Failing to reject H₀ when H₁ is true (false negative).  
Example:  
Type I: Diagnosing a healthy person as sick.  
Type II: Missing a disease in a sick person.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?  
- One-tailed test: Tests for effect in one direction (e.g., greater than).  
- Two-tailed test: Tests for effect in both directions (e.g., not equal to).  
Example:  
One-tailed: H₁: μ > 50  
Two-tailed: H₁: μ ≠ 50

8. What is the Z-test, and when is it used in hypothesis testing?  
- A Z-test is used when the population variance is known and the sample size is large (n > 30). It compares sample and population means.  
Example:  
Testing if average height in a city differs from national average using known σ.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?  
- Z = (X̄ - μ) / (σ / √n)  
- The Z-score tells how many standard deviations the sample mean is from the population mean.  
Example:  
If Z = 2, the sample mean is 2 standard deviations above the population mean.

10. What is the T-distribution, and when should it be used instead of the normal distribution?  
- The T-distribution is used when the sample size is small (n < 30) and population standard deviation is unknown.  
- It has heavier tails than the normal distribution, accounting for more uncertainty.  
Example:  
Using student test scores (n = 15) without knowing population σ → Use t-distribution.

11. What is the difference between a Z-test and a T-test?  
- Z-test: Used when population standard deviation is known and sample size is large (n > 30).  
- T-test: Used when population standard deviation is unknown and/or sample size is small (n < 30).  
Example:  
Z-test for height comparison with known σ, T-test for test scores with unknown σ.

12. What is the T-test, and how is it used in hypothesis testing?  
- T-test compares the means of one or two groups to determine if they are significantly different.  
- It uses the T-distribution when σ is unknown.  
Example:  
One-sample t-test to check if class average score ≠ 75.

13. What is the relationship between Z-test and T-test in hypothesis testing?  
- Both test hypotheses about population means.  
- Z-test is used with known σ and large samples; T-test is used with unknown σ and small samples.  
- As sample size increases, T-distribution approaches normal distribution (Z).

14. What is a confidence interval, and how is it used to interpret statistical results?  
- A confidence interval gives a range of values likely to contain the true population parameter.  
- It’s calculated from sample data and associated with a confidence level (e.g., 95%).  
Example:  
CI = [68, 72] with 95% confidence → population mean lies between 68 and 72.

15. What is the margin of error, and how does it affect the confidence interval?  
- Margin of error is the maximum expected difference between the sample estimate and true population value.  
- Larger margin → wider confidence interval.  
- Depends on sample size and confidence level.  
Example:  
Mean = 70, Margin of Error = 2 → CI = [68, 72]

16. How is Bayes' Theorem used in statistics, and what is its significance?  
- Bayes' Theorem updates probability estimates based on new evidence.  
- It's used in decision-making, medical testing, machine learning, etc.  
Formula:  
P(A|B) = [P(B|A) * P(A)] / P(B)  
Example:  
Used to revise the probability of having a disease after a positive test result.

17. What is the Chi-square distribution, and when is it used?  
- Chi-square distribution is used for categorical data and variance analysis.  
- It arises when summing squared standard normal variables.  
Example:  
Used in tests like Chi-square goodness of fit or independence tests.

18. What is the Chi-square goodness of fit test, and how is it applied?  
- It checks if observed frequencies match expected frequencies in categorical data.  
- Helps determine if a sample follows a specified distribution.  
Example:  
Testing if a die is fair by comparing observed vs expected roll outcomes.

19. What is the F-distribution, and when is it used in hypothesis testing?  
- F-distribution is used to compare two variances.  
- It is right-skewed and used in ANOVA and F-tests.  
Example:  
Used in testing if variability in different groups is equal.

20. What is an ANOVA test, and what are its assumptions?  
- ANOVA (Analysis of Variance) checks if means of three or more groups are significantly different.  
Assumptions:  
  - Samples are independent  
  - Populations are normally distributed  
  - Equal variances across groups  
Example:  
Comparing test scores across different teaching methods.

21. What are the different types of ANOVA tests?  
- One-way ANOVA: Compares means of groups based on one factor.  
- Two-way ANOVA: Compares means based on two factors and their interaction.  
- Repeated Measures ANOVA: For related groups or time-based comparisons.  
Example:  
One-way ANOVA for 3 classes' scores, Two-way ANOVA for teaching method & gender.

22. What is the F-test, and how does it relate to hypothesis testing?  
- F-test compares two variances to see if they are significantly different.  
- It’s also used in ANOVA to compare group variances.  
Example:  
Used in regression analysis to test if a model is a good fit.

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

import numpy as np
from scipy.stats import norm

# Sample data
data = np.array([105, 110, 102, 108, 100, 112, 107])
sample_mean = np.mean(data)
pop_mean = 100
pop_std = 5
n = len(data)

# Z-score
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

# P-value
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-score: {z:.3f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
# 2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python

import numpy as np
from scipy.stats import ttest_1samp

# Simulate data
np.random.seed(0)
data = np.random.normal(loc=50, scale=5, size=100)

# Perform one-sample t-test
t_stat, p_value = ttest_1samp(data, popmean=52)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")
```

```python
# 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean

import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, pop_mean, pop_std):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))
    return z_score, p_value

data = np.random.normal(loc=49, scale=5, size=50)
z, p = one_sample_z_test(data, pop_mean=50, pop_std=5)

print(f"Z-score: {z:.3f}")
print(f"P-value: {p:.3f}")
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
sample_mean = 102
pop_mean = 100
pop_std = 5
n = 30

# Z-score and p-value
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p = 2 * (1 - norm.cdf(abs(z)))

# Visualization
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)
critical_z = norm.ppf(0.975)  # 95% confidence

plt.plot(x, y, label='Z-distribution')
plt.fill_between(x, y, where=(x < -critical_z) | (x > critical_z), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(z, color='black', linestyle='--', label=f'Z-score = {z:.2f}')
plt.legend()
plt.title("Two-tailed Z-test")
plt.xlabel("Z")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()

print(f"Z-score: {z:.3f}")
print(f"P-value: {p:.3f}")
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2(mu0=100, mu1=102, sigma=5, n=30, alpha=0.05):
    se = sigma / np.sqrt(n)
    x = np.linspace(90, 110, 1000)

    # Distributions under H0 and H1
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    # Critical value for Type I error
    z_crit = norm.ppf(1 - alpha)
    crit_val = mu0 + z_crit * se

    plt.plot(x, y0, label='H0 distribution', color='blue')
    plt.plot(x, y1, label='H1 distribution', color='green')
    plt.fill_between(x, y0, where=(x >= crit_val), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, y1, where=(x < crit_val), color='orange', alpha=0.3, label='Type II Error')
    plt.axvline(crit_val, color='black', linestyle='--', label=f'Critical Value = {crit_val:.2f}')
    plt.title("Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_type1_type2()
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    stderr = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)
    margin = t_crit * stderr
    return mean - margin, mean + margin

sample = np.random.normal(50, 5, 40)
ci = confidence_interval(sample)
print(f"Confidence Interval (95%): {ci}")
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior_A, prob_B_given_A, prob_B):
    return (prob_B_given_A * prior_A) / prob_B

# Example: disease testing
prior_disease = 0.01       # P(Disease)
prob_positive_given_disease = 0.99  # P(Positive | Disease)
prob_positive = 0.05       # P(Positive overall)

posterior = bayes_theorem(prior_disease, prob_positive_given_disease, prob_positive)
print(f"Posterior Probability (P(Disease | Positive)): {posterior:.4f}")

```

```python
# 16. Implement an F-test using Python to compare the variances of two random samples

import numpy as np
from scipy.stats import f

sample1 = np.random.normal(10, 2, 50)
sample2 = np.random.normal(10, 3, 50)

var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

f_stat = var1 / var2 if var1 > var2 else var2 / var1
df1, df2 = 49, 49
p_value = 1 - f.cdf(f_stat, df1, df2)

print(f"F-statistic: {f_stat:.3f}")
print(f"P-value: {p_value:.3f}")
```

```python
# 17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results

import numpy as np
from scipy.stats import f_oneway

g1 = np.random.normal(50, 10, 30)
g2 = np.random.normal(52, 10, 30)
g3 = np.random.normal(55, 10, 30)

f_stat, p_val = f_oneway(g1, g2, g3)

print(f"F-statistic: {f_stat:.3f}")
print(f"P-value: {p_val:.3f}")
```

```python
# 18. Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results

import matplotlib.pyplot as plt

groups = [g1, g2, g3]
plt.boxplot(groups, labels=['Group 1', 'Group 2', 'Group 3'])
plt.title("Boxplot of Groups for One-Way ANOVA")
plt.ylabel("Values")
plt.grid()
plt.show()
```

```python
# 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import scipy.stats as stats

def check_anova_assumptions(*groups):
    for i, group in enumerate(groups):
        stat, p = stats.shapiro(group)
        print(f"Group {i+1} Shapiro-Wilk Normality Test: p-value = {p:.3f}")

    stat, p = stats.levene(*groups)
    print(f"Levene’s Test for Equal Variances: p-value = {p:.3f}")

check_anova_assumptions(g1, g2, g3)
```

```python
# 20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

data = pd.DataFrame({
    'Score': np.random.normal(50, 10, 60),
    'Gender': ['Male']*30 + ['Female']*30,
    'Treatment': ['A']*15 + ['B']*15 + ['A']*15 + ['B']*15
})

model = ols('Score ~ C(Gender) + C(Treatment) + C(Gender):C(Treatment)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
# 23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means

sample = np.random.normal(50, 5, 100)
t_stat, p_val = stats.ttest_1samp(sample, popmean=50)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_val:.3f}")
```

```python
# 24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

n = 30
data = np.random.normal(0, 2, n)
sample_var = np.var(data, ddof=1)
hyp_var = 4

chi2_stat = (n - 1) * sample_var / hyp_var
p_val = 1 - stats.chi2.cdf(chi2_stat, df=n-1)

print(f"Chi-square Statistic: {chi2_stat:.3f}")
print(f"P-value: {p_val:.3f}")
```

```python
# 25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

from statsmodels.stats.proportion import proportions_ztest

successes = np.array([40, 30])
samples = np.array([100, 100])

z_stat, p_val = proportions_ztest(successes, samples)
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_val:.3f}")
```

```python
# 26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

data1 = np.random.normal(0, 2, 100)
data2 = np.random.normal(0, 3, 100)

f_stat = np.var(data1, ddof=1) / np.var(data2, ddof=1)
df1 = len(data1) - 1
df2 = len(data2) - 1

p_val = 1 - stats.f.cdf(f_stat, df1, df2)

print(f"F-statistic: {f_stat:.3f}")
print(f"P-value: {p_val:.3f}")
```
